# Customer Churn Prediction: Exploratory Data Analysis

## 1. Business Understanding
Churn prediction is critical because acquiring a new customer is often much more expensive than retaining an existing one. By predicting churn, we can proactively engage high-risk customers with targeted retention strategies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

data_path = '../WA_Fn-UseC_-Telco-Customer-Churn.csv'
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
else:
    print('Dataset not found at expected location.')
df.head()
```

```python
# 2. Missing Values & Data Types
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
df.info()
```

```python
# 3. Target Variable Distribution
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```

```python
# 4. Feature vs Churn Relationship
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tenure vs Churn Risk')
plt.show()
```

```python
# 5. Contract vs Churn Relationship
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Contract Type vs Churn Risk')
plt.show()
```

**Business Insight**: 
1. Customers with lower tenure have a significantly higher risk of churning. Early onboarding engagement is critical.
2. Month-to-month contracts are highly susceptible to churn compared to one/two year contracts.
